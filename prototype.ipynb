```python
 def sign_message(msg: str, secret: int, pubkey: PublicKey) -> Tuple[int, int]:
    """Sign message based on Schnorr's signature"""
    k = secrets.randbits(2048)
    r = pow(pubkey.g, k, pubkey.p)
    print("r = ", r, "\n")
    e = hash_as_int(str(r) + msg) % pubkey.q
    s = (k  - secret * e) % pubkey.q
    print("s = ", s, "\n")
    print("e = ", e, "\n")
    return s, e

def verify_message(msg: str, s: int, e: int, pubkey: PublicKey) -> bool:
    r_v = pow(pubkey.g, s, pubkey.p) * pow(pubkey.g_x, e, pubkey.p) % pubkey.p
    e_v = hash_as_int(str(r_v) + msg) % pubkey.q
    print("r_v = ", r_v, "\n")
    print("e_v = ", e_v, "\n")
    return e == e_v
```
